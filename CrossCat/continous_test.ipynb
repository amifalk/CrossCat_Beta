```python
from crosscat import CrossCat
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import default_rng
```

```python
top = default_rng().normal(loc=0, scale=1, size=(5,3))
middle = default_rng().normal(loc=10, scale=1, size=(5,3))
bottom = default_rng().normal(loc=-10, scale=1, size=(5,3))

data = np.concatenate((top, middle, bottom), axis=0)
```

```python
plt.matshow(data)
```

```python
model = CrossCat(data)
```

```python
model.curr.plot()
```

```python
model.run(10)
```

```python
model.best_fit().plot()
```

```python
model.plot_liks()
```
